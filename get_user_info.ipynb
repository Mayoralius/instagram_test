```python
import os
import requests
import json
from datetime import datetime
```

```python
headers = json.loads(open('./config.json', 'r').read())['request']['headers']
```

```python
def get_user_info(user):
    return requests.get('https://i.instagram.com/api/v1/users/web_profile_info/',
            params={ 'username': user },
            headers=headers
    ).json()
```

```python
def request_instagram(max_id=0):
    return requests.get('https://i.instagram.com/api/v1/friendships/46439992559/followers/',
                 params={
                    'search_surface': 'follow_list_page',
                    'max_id': max_id,
                 },
                 headers=headers).json()
```

```python
user = get_user_info('martabermar_')
filename = '%s.json' %user['data']['user']['username']
```

```python
user_followers_count = user['data']['user']['edge_followed_by']['count']
user_following_count = user['data']['user']['edge_follow']['count']
```

```python
followers = {}
max_tries = 0
while len(followers) < user_followers_count and max_tries < 3:
    max_id = 0
    for max_id in range(0, user_followers_count, 100):
        new_users = request_instagram(max_id)
        max_tries += 1
        print(len(new_users['users']))
        for user in new_users['users']:
            if user['username'] not in followers:
                followers[user['username']] = user
```

```python
old_followers = {}
if filename in os.listdir('./users'):
    with open('./users/%s' %filename, 'r') as file:
        old_followers = json.loads(file.read())
        file.close()
```

```python
new_followers = {}
timestamp = datetime.now()
for k, v in followers.items():
    if k not in old_followers.keys():
        new_followers[k] = v
        new_followers[k]['starts_following'] = timestamp
print('%d new followers detected' %len(new_followers))
if len(new_followers):
    old_followers.update(new_followers)
    with open('./users/%s' %filename, 'w') as file:
        file.truncate()
        file.write(json.dumps(old_followers, default=str))
        file.close()
```
